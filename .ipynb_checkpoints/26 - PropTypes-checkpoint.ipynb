# 26: PropTypes


#### Create a new React project called *PropTypes*
Open **Command Prompt** on Windows or **Terminal** on Mac or Linux and enter the following to create a new React project named *PropTypes*:

```
create-react-app PropTypes
```

This starts the **create-react-app** program, which will create a new directory named *PropTypes*, and build a brand new project inside that directory.

Once the project has been created, go into the *PropTypes* directory. Do this by entering the following:

```
cd PropTypes
```

> If you’re working the command line and need to know which directory you’re in, simply enter `cd` on Windows or `pwd` on Mac or Linux.


#### Install the *prop-types* package

Once you’re in the *PropTypes* directory, install the *prop-types* package by entering:

```
npm install prop-types
```

This tells **npm**, the Node Package Manager, to install a package named *prop-types* into the project. Once installed, any JavaScript file in the project can make use of the *prop-types* package.


#### Replace the code in *src/App.js*

Open the project’s *App.js* file (it’s in the *src* directory) and replace its contents with the following:

```
import './App.css';

import PropTypes from 'prop-types';
function Comment({ author, message, likes, children }) { 
  return (
    <div>
      <div className='author'>{author}</div> 
      <div className='message'>{message}</div> 
      <div className='likes'>
        {likes > 0 ? likes : 'No'} likes 
      </div>
    </div>
  );
}

Comment.propTypes = {
  message: PropTypes.string.isRequired, 
  author: PropTypes.string.isRequired, 
  likes: PropTypes.number
}

function App() {
  return (
    <div className="App">
      <Comment author='somebody' message='a likable message' likes={1}/>
    </div>
  );
}

export default App;
```


#### Run the project and open the browser console

Save the changes you made to *App.js* and run the project by entering the following in the command line:

```
npm start
```

Open the browser console. You should not see any error messages.
