# 02: Installing the necessities for Part 1 of the course

## Node.js and npm

You’ll need to install [**Node.js**](https://nodejs.org/) and [**npm**](https://www.npmjs.com/); pretty much all the work in the course will rely on them.

Before you install these applications, here’s a quick explanation of what they are.

[Go to the the *Downloads* page on the Visual Studio Code site](https://code.visualstudio.com/download), download the installer. Double-click on the installer file to install Visual Studio code.


### What is Node.js?

<img src="http://www.globalnerdy.com/wp-content/uploads/2017/04/node.js-logo.jpg" align="right" width="150" style="padding: 10px;" alt="Node.js logo" />

[**Node.js**](https://nodejs.org/) (pronounced “Node J S” or simply “Node”) is an engine that allows you to run JavaScript outside of the browser. Priort to Node, JavaScript engines existed only in browsers, and as a result, JavaScript programs could only exist and run inside web pages.

Node is built on [V8](https://v8.dev/), the efficient and powerful JavaScript engine inside the Chrome browser, but instead of running inside a browser, it runs inside a computer operating system, such as Windows, macOS, and Linux. With Node, JavaScript graduated from a webpage-only programming language to a general purpose programming language.

In this course, you’ll use Node to:

* Enter short JavaScript commands and get immediate feedback
* Write programs that you’ll run on your computer
* Write the “back end” of web applications


### What is npm?

<img src="http://www.globalnerdy.com/wp-content/uploads/2020/09/npm-logo.jpg" align="right" width="150" style="padding: 10px;" alt="npm logo" />

[**npm**](https://www.npmjs.com/) is short for Node Package Manager. It’s an application that lets you download and install pre-written packages of JavaScript code (think of them as “plugins for programmers”) for use in your own JavaScript programs. By using these packages, you’re saved from having to “reinvent the wheel”.

In this course, you’ll use npm to:

* Download and install modules of pre-written code to use in your own JavaScript programs.


### Installing Node.js and npm on Windows

[Go to the the *Downloads* page on the Node.js site](https://nodejs.org/en/download/) and download the appropriate **Windows Installer (.msi)** file for your system — it’ll be either the 32-bit edition or the 64-bit edition. Double-click on the installer file to install Node.js and npm, which comes bundled with Node.

> If you don’t know if your Windows is 32- or 64-bit, [follow these steps.](https://support.microsoft.com/en-us/help/13443/windows-which-version-am-i-running) 


### Installing Node.js and npm on macOS

[Go to the the *Downloads* page on the Node.js site](https://nodejs.org/en/download/) and download the **Mac Installer (.pkg)** file. Double-click on the installer file to install Node.js and npm, which comes bundled with Node.


### Installing Node.js and npm on an Ubuntu-based or Debian-based Linux distribution

If your flavor of Linux uses the `apt` (formerly known as `apt-get`) to install applications, it’s probably Ubuntu, Debian, or some Linux distribution based on Ubuntu or Debian.

For these flavors of Linux, open a terminal and enter the following to ensure that the `apt` application (which you’ll use to install Node) has the most up-to-date list of sources of software:

```
sudo apt update
```

You’ll be asked for your password in the process.

Then enter the following to install Node:

```
sudo apt install nodejs
```

Once Node has been installed, you’ll need to install npm. Enter the following to do so:

```
sudo apt install npm
```

## Visual Studio Code

<img src="http://www.globalnerdy.com/wp-content/uploads/2020/07/visual-studio-code.jpg" src="http://www.globalnerdy.com/wp-content/uploads/2020/07/visual-studio-code.jpg" align="center" width="200" style="padding: 10px;" alt="Visual Studio Code logo" />

You’ll also need to install **Visual Studio Code**, which is a text editor. Think of it as “Notepad on steroids”. It’s made specifically for entering and editing code, and thanks to its large library of plugins, supports many programming languages, JavaScript included.


### Installing Visual Studio Code on Windows

[Go to the the *Downloads* page on the Visual Studio Code site](https://code.visualstudio.com/download) and download the appropriate **User Installer** file for your system — it’ll be either the 32-bit edition or the 64-bit edition. Double-click on the installer file to install Visual Studio code.

> If you don’t know if your Windows is 32- or 64-bit, [follow these steps.](https://support.microsoft.com/en-us/help/13443/windows-which-version-am-i-running) 


### Installing Visual Studio Code on macOS

[Go to the the *Downloads* page on the Visual Studio Code site](https://code.visualstudio.com/download) and download the installer. Double-click on the installer file to install Visual Studio code.


### Installing Visual Studio Code on an Ubuntu-based or Debian-based Linux distribution

Open a terminal and enter the following to ensure that you have the `wget` application by entering the following into a terminal:

```
sudo apt install software-properties-common apt-transport-https wget
```

Use the `wget` application to retrieve the Microsoft key required to retrieve Visual Studio Code by entering the following:

```
wget -q https://packages.microsoft.com/keys/microsoft.asc -O- | sudo apt-key add -
```

Enable the Microsoft repository by entering this:

```
sudo add-apt-repository "deb [arch=amd64] https://packages.microsoft.com/repos/vscode stable main"
```

And finally, enter this to download and install Visual Studio Code:

```
sudo apt install code
```
