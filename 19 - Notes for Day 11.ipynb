## Weather app

### The code

```
<html>

<head>
<script>
// Replace XXXXXXX with your OpenWeatherMap API key.
const requestURL = `http://api.openweathermap.org/data/2.5/weather?q=Paris,TX,US&APPID=XXXXXXX`

// The fetch() function takes an URL and 
// returns the information from that URL.
fetch(requestURL)
  .then(response => response.json())
  .then(json => displayWeather(json))

// displayWeather() takes the information returned from
// OpenWeatherMap and displays it on the web page.
function displayWeather(data) {
  console.log(data)
  console.log(`data.main`)
  console.log(data.main)
  console.log(`data.wind`)
  console.log(data.wind)

  document.getElementById("temp_f").innerHTML = Math.round(kelvinToFahrenheit(data.main.temp))
  document.getElementById("temp_c").innerHTML = Math.round(kelvinToCelsius(data.main.temp))
  document.getElementById("wind_speed").innerHTML = Math.round(metersPerSecondToMilesPerHour(data.wind.speed))
  document.getElementById("sunrise").innerHTML = data.sys.sunrise
  document.getElementById("sunset").innerHTML = data.sys.sunset

}


// Conversion functions
// --------------------
// These functions take the data provided by OpenWeatherMap
// and convert them into formats that are more common,
// or easier to read or understand.

function kelvinToFahrenheit(kelvin) {
  return kelvinToCelsius(kelvin) * 9/5 + 32
}

function kelvinToCelsius(kelvin) {
  return (kelvin - 273.15)
}

function metersPerSecondToMilesPerHour(metersPerSecond) {
  return metersPerSecond * 2.237
}
</script>
</head>

<body>
  <p>Temperature: <span id="temp_f"></span>F <span id="temp_c"></span>C</p>
  <p>Wind speed: <span id="wind_speed"></span> mph</p>
  <p>Sunrise: <span id="sunrise"></span></p>
  <p>Sunset: <span id="sunset"></span></p>
</body>

</html>
```

### The code, explained

#### 1. Where are we getting the weather information from?

We’re getting it from the [OpenWeatherMap](https://openweathermap.org/) service, which collects data from weather stations all over the world, and makes that data available through an **API**.


#### 2. What’s an API?

“API” is short for ***Application Programming Interface***. It’s the set of services offered by another computer that your program can use. 

A good analogy for APIs is a restaurant menu. You choose the item you want, and if it’s possible, you can even specify some options. Then you place the order. Some time after that, you get your item.

APIs, like restaurant menus, provide a set of choices. You choose the item you want, and if it’s possible, you can even specify some options. Then you send your order, which is called a **request**, to the system with the API. Some time after that, you’ll get something back from that system: the **response**.

OpenWeatherMap has an API that lets you get the current weather for any of over 200,000 locations worldwide. You provide it with information that lets it identify the location — in this example, we’ll use the name of a city or town — and it will respond with data describing the current weather at that location.
