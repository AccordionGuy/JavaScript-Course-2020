# 20: The Weather app and the Star Wars app

## The Weather app

### Code

```
<html>

<head>
<script>
function getWeather(place) {
  const trimmedPlace = place.trim()
  if (trimmedPlace.length > 0) {
    const requestURL = `http://api.openweathermap.org/data/2.5/weather?q=${trimmedPlace}&APPID=06db44f389d2172e9b1096cdce7b051c`
    fetch(requestURL)
      .then(response => response.json())
      .then(json => {
        if (json.cod != "404") {
          displayWeather(json)
        } else {
          displayCityNotFound()
        }
      })
  } else {
    alert("Please enter a city name.")
  }
}

function displayCityNotFound() {
  alert("I don't recognize that city name.")
  const placeTextBox = document.getElementById("place")
  placeTextBox.value = ""
  placeTextBox.focus()
}

function displayWeather(data) {
  console.log(data)
  console.log(`data.main`)
  console.log(data.main)
  console.log(`data.wind`)
  console.log(data.wind)

  document.getElementById("forecast").style.visibility = "visible"

  document.getElementById("name").innerHTML = data.name
  document.getElementById("temp_f").innerHTML = Math.round(kelvinToFahrenheit(data.main.temp))
  document.getElementById("temp_c").innerHTML = Math.round(kelvinToCelsius(data.main.temp))
  document.getElementById("wind_speed").innerHTML = Math.round(metersPerSecondToMilesPerHour(data.wind.speed))
  document.getElementById("wind_description").innerHTML = windSpeedToDescription(data.wind.speed)
  document.getElementById("wind_direction").innerHTML = degreesToDirection(data.wind.deg)
  const sunriseTime = new Date(data.sys.sunrise * 1000)
  const sunsetTime = new Date(data.sys.sunset * 1000)
  document.getElementById("sunrise").innerHTML = sunriseTime
  document.getElementById("sunset").innerHTML = sunsetTime

  const weatherDescription = data.weather[0].description
  const weatherIcon = data.weather[0].icon
  const weatherId = data.weather[0].id
  const weatherMain = data.weather[0].main

  document.getElementById("description").innerHTML = weatherDescription
  document.getElementById("icon").src = `http://openweathermap.org/img/wn/${weatherIcon}@2x.png`
}

function kelvinToFahrenheit(kelvin) {
  return kelvinToCelsius(kelvin) * 9/5 + 32
}

function kelvinToCelsius(kelvin) {
  return (kelvin - 273.15)
}

function metersPerSecondToMilesPerHour(metersPerSecond) {
  return metersPerSecond * 2.237
}

function degreesToDirection(degrees) {
  const CardinalDirections = ["N", "NNE", "NE", "ENE", "E", "ESE", "SE", "SSE", "S", 
                              "SSW", "SW", "WSW", "W", "WNW", "NW", "NNW"]
  const NumberOfSectors = CardinalDirections.length
  const DegreesPerSector = 360 / NumberOfSectors

  const sector = Math.floor((degrees / DegreesPerSector) + 0.5)
  return CardinalDirections[(sector % NumberOfSectors)]
}

function metersPerSecondToKnots(metersPerSecond) {
  return metersPerSecond * 1.944
}

function windSpeedToDescription(windSpeed) {
  const Descriptions = [
    "Calm",
    "Light air",
    "Light breeze",
    "Gentle breeze",
    "Moderate breeze",
    "Fresh breeze",
    "Strong breeze",
    "Moderate gale",
    "Fresh gale",
    "Strong gale",
    "Whole gale",
    "Violent storm",
    "Hurricane"
  ]

  const speed = metersPerSecondToKnots(windSpeed)

  let scale = 0
  if (speed >= 1 && speed <= 3) {
    scale = 1
  } else if (speed <= 7) {
    scale = 2
  } else if (speed <= 12) {
    scale = 3
  } else if (speed <= 18) {
    scale = 4
  } else if (speed <= 24) {
    scale = 5
  } else if (speed <= 31) {
    scale = 6
  } else if (speed <= 38) {
    scale = 7
  } else if (speed <= 46) {
    scale = 8
  } else if (speed <= 54) {
    scale = 9
  } else if (speed <= 63) {
    scale = 10
  } else if (speed <= 72) {
    scale = 11
  } else {
    scale = 12
  }

  return Descriptions[scale]
}


</script>
</head>

<body style="background-color:powderblue;">
  <p>
    <label for="place">Enter a city: </label>
    <input id="place" type="text">
    <button onclick="getWeather(document.getElementById('place').value)">Get weather</button>
  </p>

<div id="forecast" style="visibility: hidden">
  <h2>Forecast for <span id="name"></span></h2>
  <p>Temperature: <span id="temp_f"></span>F <span id="temp_c"></span>C</p>
  <p>Wind speed: <span id="wind_speed"></span> mph (<span id="wind_description"></span>)</p>
  <p>Wind direction: <span id="wind_direction"></span></p>
  <p>Sunrise: <span id="sunrise"></span></p>
  <p>Sunset: <span id="sunset"></span></p>

  <p><strong>Weather description:</strong> <span id="description"></span></p>
  <p><img id="icon" /></p>
</div>

</body>

</html>
```
