```javascript
for (number = 1; number <= 100; number++) {
  word = ""
  const isMultipleOf3 = (number % 3 == 0)
  const isMultipleOf5 = (number % 5 == 0)

  if (isMultipleOf3 || isMultipleOf5) {
    if (isMultipleOf3) {
      word += "Fizz"
    }
    if (isMultipleOf5) {
      word += "Buzz"
    }
  } else {
    word = number.toString()
  }
  console.log(word)
}
```
